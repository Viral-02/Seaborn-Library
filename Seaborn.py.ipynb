```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

#### Correlation

```python
df.corr()
```

##### Heatmap of Correlation

Plot rectangular data as a color-encoded matrix.A heatmap contains values representing various shades of the same colour for each value to be plotted. Below we have plotted the correlation of all the columns which has numerical values. It shows the relation between each column to other columns. We can refer to the range line at the right side of the heatmap to understnad how strongly or weakly the correlation is.

```python
sns.heatmap(df.corr())
```

#### Joint Plot

A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, 2D histogram or 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### Pair Plot
Pair Plots are a really simple (one-line-of-code simple!) way to visualize relationships between each variable. It produces a matrix of relationships between each variable in your data for an instant examination of our data. It can also be a great jumping off point for determining types of regression analysis to use.

```python
sns.pairplot(df)
```

Below we have added 'hue' parameter
Variable in data to map plot aspects to different colors.

```python
sns.pairplot(df,hue='sex')
```

#### DistPlot
A distplot plots a univariate distribution of observations. 
The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

#### Countplot

Countplot can be viewed as histogram of Categorical values. It gives count of each type in the categorical variable. 

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='day',data=df)
```

#### Bar Plot
It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

#### Box Plot
A box and whisker plot(sometimes called a boxplot) is a graph that presents information from a five-number summary:
The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile.

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

#### Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel Density Estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,pallete='rainbow')
```
